```python
subs_contrat
```

Les trois IN ci-haut comportent les étapes que j'ai suivi pour traduire les colonnes restantes en français. Après les avoir affichées, j'ai changé les termes en français.

```python
subs_contrat["valeur_entente"] = subs_contrat.valeur_entente.str.replace("$", "").str.replace(",","")
```

```python
subs_contrat["valeur_entente"] = subs_contrat.valeur_entente.astype(float)
```

```python
subs_contrat
```

Avec l'aide du web : https://www.ritchieng.com/pandas-changing-datatype/ , j'ai réussi à changer ma données "valeur_entente" en données lisible par Pandas. Pour ce faire, j'ai remplacé les symboles $ et , par des "" (donc rien) et ensuite j'ai changé le type de la variable, un peu comme nous faisions dans python avec le int() ou str() par exemple.
